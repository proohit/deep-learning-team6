```python

import os
import cv2

for i in range (0,105): 
    # create folder to store extracted images
    folder = 'videoframes/all/Frames_video'+ str(i)
    os.mkdir(folder)
    #extract video
    vidcap = cv2.VideoCapture('/home/dl6/deep-learning-team6/1b/trainingvideos/_8Vy3dlHg2w_00' + str(i).zfill(3)+ '.mp4')
    count = 0
    while True:
        success,image = vidcap.read()
        if not success:
            break
        cv2.imwrite(os.path.join(folder,"frame{:d}.jpg".format(count)), image)  
        count += 1
    print(str(count) + ' images are extacted in video ' + str(i))
```

```python
videoframe_num={}
size_all = 0
  
for i in range(0,105): 
    traindata= os.listdir('/home/dl6/deep-learning-team6/1b/videoframes/all/Frames_video'+ str(i))
    
    videoframe_num['video'+str(i)]=len(traindata)
   
    size_all+=len(traindata)
    
    print('video ' +str(i)+ ' has ' + str(videoframe_num['video'+str(i)]) + ' images')
    
print('sum of images:' + str(size_all))
```

```python
import matplotlib.pyplot as plt
df = pd.DataFrame.from_dict(videoframe_num, orient='index')

plt.boxplot(df)
plt.show
```

```python
size_side = 0
size_front = 0
size_downwards = 0
size_upwards = 0

size_folder = 0

avg_side=0
avg_upwards=0
avg_downwards=0
avg_front=0

# Upwards
for (root,dirs,files) in os.walk('/home/dl6/deep-learning-team6/1b/videoframes/Upwards/'):
    
    size_folder+=len(dirs)
    size_upwards+=len(files)

avg_upwards= size_upwards / size_folder
print('Kameraposition: Upwards, '+ ' size of images: ' + str(size_upwards)+ ', size of folders: '+ str(size_folder))


# Side
size_folder = 0
for (root,dirs,files) in os.walk('/home/dl6/deep-learning-team6/1b/videoframes/Side/'):
    
    size_folder+=len(dirs)
    size_side+=len(files)
    
avg_side= size_side / size_folder
    
print('Kameraposition: Side, ''size of images: ' + str(size_side)+ ', size of folder: '+ str(size_folder))

# Downwards
size_folder = 0
for (root,dirs,files) in os.walk('/home/dl6/deep-learning-team6/1b/videoframes/Downwards/'):
    
    size_folder+=len(dirs)
    size_downwards+=len(files)

avg_downwards= size_downwards / size_folder
print('Kameraposition: Downwards, ''size of images: ' + str(size_downwards)+ ', size of folder:  '+ str(size_folder))

# Front
size_folder = 0
for (root,dirs,files) in os.walk('/home/dl6/deep-learning-team6/1b/videoframes/Front/'):
    
    size_folder+=len(dirs)
    size_front+=len(files)
    
avg_front= size_front / size_folder

print('Kameraposition: Front, ''size of images: ' + str(size_front)+ ', size of folder:  '+ str(size_folder))

# Boxplotdiagramm / Pyplot  erstellen oder Durchschittsvergleich mit andere Ordner 
```

```python
names = [ 'upwards', 'downwards','front', 'side']
values = [ size_upwards, size_downwards,size_front, size_side]


plt.figure(figsize=(20, 3))
plt.subplot(131)
plt.bar(names, values)
plt.title('Anzahl Bilder pro Kameraposition')
plt.show()

names = [ 'upwards', 'downwards','front', 'side']
values = [ avg_upwards, avg_downwards,avg_front, avg_side]


plt.figure(figsize=(20, 3))
plt.subplot(131)
plt.bar(names, values)
plt.title('Durchschnittsanzahl Bilder pro Kameraposition')
plt.show()
```
